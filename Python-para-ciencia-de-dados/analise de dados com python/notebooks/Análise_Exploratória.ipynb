```python
# Criando nosso DataFrame
df = pd.read_excel("/content/drive/MyDrive/datasets/AdventureWorks.xlsx")
```

```python
# Visualiando as 5 primeiras linhas
df.head(5)
```

```python
# Quantidade de linhas e colunas
df.shape
```

```python
# Verificando os tipo de dados
df.dtypes
```

```python
# Qual a receita total?
df['Valor Venda'].sum()
```

```python
# Qual o custo Total?
df["Custo"] = df["Custo Unitário"].mul(df["Quantidade"]) # Criando a coluna de custo
```

```python
df.head(2)
```

```python
# Qual o custo total?
round(df['Custo'].sum(), 2)
```

```python
# Agora que temos a receita e o custo total, podemos achar o lucro total
# Vamos criar uma coluna de lucro que será a receita - custo

df["lucro"] = df["Valor Venda"] - df["Custo"]
df.sample(5)
```

```python
# Total Lucro
round(df["lucro"].sum(), 2)
```

```python
# Criando uma coluna com total de dias para enviar o produto
df["Tempo_envio"] = df["Data Envio"] - df['Data Venda']
df.head(5)
```

## Agora queremos saber a média de tempo de envio para cada Marca e para isso precisamos transformar a coluna Tempo_envio em númerica

```python
# Extraindo apenas os dias
df["Tempo_envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
```

```python
df.head(5)
```

```python
# Verificando o tipo de coluna Tempo_envio
df["Tempo_envio"].dtypes
```

```python
# Média de tempo de envio para cada Marca
df.groupby("Marca")["Tempo_envio"].mean()
```

### Missing Values

```python
# Verificar se tempos dados faltantes
df.isnull().sum()
```

### E se a gente quiser saber o lucro por ano e por marca?


```python
# Vamos agrupar por ano e marca
df.groupby([df['Data Venda'].dt.year, "Marca"])["lucro"].sum()
```

```python
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
# Resetando index
lucro_ano = df.groupby([df['Data Venda'].dt.year, "Marca"])["lucro"].sum().reset_index()
```

```python
# Qual o total de produtos vendidos?
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False)
```

```python
# Gráfico total de produtos vendidos
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=True).plot.barh(title="Total de produtos vendidos")
plt.xlabel = "Total"
plt.ylabel = "Produto";
```

```python
df.groupby(df["Data Venda"].dt.year)["lucro"].sum().plot.bar(title="Lucro x Ano")
plt.xlabel = "Ano"
plt.ylabel = "Receita"
```

```python
df.groupby(df["Data Venda"].dt.year)["lucro"].sum()
```

```python
# Analisando apenas as vendas de 2009
df_2009 = df[df["Data Venda"].dt.year == 2009]
```

```python
df_2009.sample(5)
```

```python
df_2009.groupby(df_2009["Data Venda"].dt.month)["lucro"].sum().plot(title="Lucro por mês")
plt.xlabel("Mês")
plt.ylabel("Lucro");
```

```python
df_2009.groupby("Marca")["lucro"].sum().plot.bar(title="Lucro x Marca")
plt.xlabel = "Marca"
plt.ylabel = "Lucro"
plt.xticks(rotation="horizontal");

```

```python
df_2009.groupby("Classe")["lucro"].sum().plot.bar(title="Classe x Marca")
plt.xlabel = "Classe"
plt.ylabel = "Lucro"
plt.xticks(rotation="horizontal");
```
